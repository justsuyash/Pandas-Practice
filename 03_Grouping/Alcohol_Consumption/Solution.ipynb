# Ex - GroupBy

### Introduction:

GroupBy can be summarized as Split-Apply-Combine.
### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from 00_Data Folder.
##### Datasets
##### Drinks - https://www.kaggle.com/datasets/mysarahmadbhat/alcohol-consumption 
##### Continents - https://www.kaggle.com/andradaolteanu/countries-iso-codes-and-population/data

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('/Users/justsuyash/Documents/GitHub/pandas_exercises/00_Data/drinks.csv')
drinks.head()
```

### Step 4. Which continent drinks more beer on average?

```python
### To do this first you have to join the the continent data set with drinks.
### Read the dataset
continents = pd.read_csv('/Users/justsuyash/Documents/GitHub/pandas_exercises/00_Data/continents.csv')
continents.head(5)
```

```python
#Use merge to join them
drinks  = pd.merge(drinks, continents[['region','name']], left_on='country',right_on='name', how='left')

#Drop Name as its not needed
drinks.drop(columns=['name'],inplace=True)
drinks.rename(columns={'region':'continents'},inplace=True)

#Check it out
drinks.head()
```

```python
drinks.groupby('continents')['beer_servings'].mean()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continents')['wine_servings'].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continents')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby('continents')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks.groupby('continents')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].agg([('Average', 'mean'), ('Maximum', 'max'), ('Minimum', 'min')])
```

```python
# Bonus:

# Calculate maximum only for beer, mean for wine and Minimum for spirit servings.
drinks.groupby('continents')[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].agg({'beer_servings': [('Maximum', 'max')], 'spirit_servings': [('Minimum', 'min')], 'wine_servings': [('Mean', 'mean')]})
```
